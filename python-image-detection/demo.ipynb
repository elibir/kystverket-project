# Image detection example: Different images
This notebook shows how to use the `Detector.py` file. We first import the `YOLOv8n` model, which can identify a bunch of objects.

```python
from detector import Detector
detector = Detector("models/yolov8n.pt")
```

In the first example we take an example of a man on a bike. Printing the detected result, we see that we find a person and a bike.

```python
detected1 = detector.find_objects_image("sample_images/random/bike.jpg")
print(detected1)
```

\\
Next we can look at a image of some animals. To get a better view of the prediction, we can plot and look at the image. this is done by setting `save_image=True`. We can also control the name of the output file by setting the `save_filename`.

```python
from PIL import Image
from IPython.display import display
detected2 = detector.find_objects_image("sample_images/random/animals.jpg",save_image=True,save_filename="pet_predictions.jpg")
print(detected2)
display(Image.open('predictions/pet_predictions.jpg'))
```

Some might notice that we only print one instance of the dogs. This has to do with our predictions only showing the max probability that their is a dog. To get an overview of all the identified objects, we set `show_all = True`. Using an example of some cars, we get:

```python
detected3 = detector.find_objects_image("sample_images/random/cars.jpg",save_image=True,save_filename="car_predictions.jpg",show_all=True)
print(detected3)
display(Image.open('predictions/car_predictions.jpg'))
```

As seen in the console, this prints all instances of cars.
